```python
import numpy as np
import json
import string
import matplotlib.pyplot as plt
import pprint
```

```python
data_path = 'Boggle/Resources/diceData.json'

with open(data_path) as json_file:
    data = json.load(json_file)
print(data)

letters_adj = [letter if letter != 'Q' else 'Qu'for letter in string.ascii_uppercase]
```

```python
letters = [dice["letters"] for dice in data]
print(letters)
```

```python
occurences = [[np.count_nonzero(np.asarray(dice) == letter) for dice in letters] for letter in letters_adj]
occurences_dict = dict(zip(letters_adj, occurences))
pprint.pprint(occurences_dict)
```

```python
probas = np.asarray([1-np.prod([(1-i/6)**val for i, val in enumerate([np.count_nonzero(np.asarray(occurence) == q) 
                                                       for q in range(0, 7)])]) for occurence in occurences])
probas = probas/sum(probas)
print(probas)
print(sum(probas))
```

```python
x_label = [letter for letter in string.ascii_uppercase]
x = np.arange(1, 27)

y_boggle = np.asarray(probas)

y_wiki_fr = [0.0745, 0.0114, 0.0324, 0.0367, 0.1444, 0.0111, 0.0123, 0.0111, 0.0664, 0.0034, 0.0029, 0.0496, 0.0262, 0.0639, 0.0507, 0.0249, 0.0065, 0.0607, 0.0651, 0.0592, 0.0454, 0.0111, 0.0017, 0.0038, 0.0046, 0.0015]
y_wiki_en = [8.167, 1.492, 2.782, 4.253, 12.702, 2.228, 2.015, 6.094, 6.966, 0.153, 0.002, 4.025, 2.406, 6.749, 7.507, 1.929, 0.095, 5.987, 6.327, 9.056, 2.758, 0.978, 2.360, 0.15, 1.974, 0.074]
y_wiki_en = np.asarray(y_wiki_en)/100
y_wiki_fr = np.asarray(y_wiki_fr)

x = np.arange(len(x_label))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(30, 10))
rects1 = ax.bar(x - width, y_boggle, width, label='Boogle')
rects2 = ax.bar(x, y_wiki_fr, width, label='Français')
rects3 = ax.bar(x + width, y_wiki_en, width, label='Anglais')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(x_label)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.1f}'.format(height*100),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.savefig('plot.jpg')
```
